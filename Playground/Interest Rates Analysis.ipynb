```python
import requests
import json
```

```python
fiscaldata_url = 'https://api.fiscaldata.treasury.gov/services/api/fiscal_service/v2/accounting/od/avg_interest_rates'
```

```python
headers = {
    'Content-Type': "application/x-www-form-urlencoded",
    'cache-control': "no-cache"
    }
```

```python
response = requests.request("GET", fiscaldata_url, headers=headers)
```

```python
response.status_code
```

```python
# from IPython.display import IFrame, HTML
# HTML(str(response.text))
```

```python
fiscaldata = json.loads(response.text)
fiscaldata
```
